```python
#Fixing columns names in the df's
allDFs[0].columns=allDFs[0].iloc[3, :].tolist()
allDFs[0].head()
```

```python
#Fixing repeated column 'HDI rank' in the first df
column_names= allDFs[0].columns.tolist()
column_names
```

```python
#How many times does 'HDI rank' appear:
count = 0

for i, col in enumerate(column_names):
    if col == 'HDI rank':
        count += 1
        #If repeated, rename
        if count == 2:
            column_names[i] = 'HDI rank.1'
            break

column_names
```

```python
allDFs[0].columns=column_names
allDFs[0].head()
```

```python
#Creating a function to clean the rows

def clean_rows(df, column_name):
        df[column_name] = pd.to_numeric(df[column_name], errors='coerce')
        cleaned_df = df.dropna(subset=[column_name])
        return cleaned_df
```

```python
#Cleaning each df using the function
allDFs = [clean_rows(df, 'HDI rank') for df in allDFs]
```

```python
#Third Problem
#Create df
df=pd.read_csv('https://github.com/UK-Data-Science/tarea3Sebastian/raw/main/fallecidosCOVID.csv')
df.head()
```

```python
#Save the previous separated columns names
c=df.columns.str.split(';')
col=c[0]
```

```python
#Separate the data
df.rename(columns={df.columns[0]:'a'},inplace=True)
df=df.a.str.split(';',expand=True)
df.info()
```

```python
#Rename the columns
newCol={0:col[0],1:col[1],2:col[2],3:col[3],4:col[4],5:col[5],6:col[6],7:col[7],8:col[8],9:col[9]}
df.rename(columns=newCol,inplace=True)
df.info()
```

```python
#Select the useful columns
keep1=['FECHA_FALLECIMIENTO']
dfSome1=df.loc[:,keep1].copy()
```

```python
#Transform the useful dates
dfSome1.FECHA_FALLECIMIENTO=pd.to_datetime(dfSome1.FECHA_FALLECIMIENTO).dt.to_period('M')
dfSome1
```

```python
#Add a count column
dfSome1.loc[:,'NUM_FALLECIDOS']=1
```

```python
#Aggregate the column
varsAggregated=['NUM_FALLECIDOS']
functionForVars='sum'
dfSome1=dfSome1.groupby('FECHA_FALLECIMIENTO')[varsAggregated].agg(functionForVars)
```

```python
#FOURTH PROBLEM
#Select the useful columns
keep1=['FECHA_FALLECIMIENTO','CLASIFICACION_DEF']
dfSome2=df.loc[:,keep1].copy()
dfSome2
```

```python
#Transform the useful dates
dfSome2.FECHA_FALLECIMIENTO=pd.to_datetime(dfSome2.FECHA_FALLECIMIENTO).dt.to_period('M')
dfSome2
```

```python
#Verify if every clasification is valid
set(dfSome2.CLASIFICACION_DEF)
```

```python
#Change 'sinadef' data
dfSome2.CLASIFICACION_DEF[dfSome2.CLASIFICACION_DEF=='sinadef']='Criterio SINADEF'
set(dfSome2.CLASIFICACION_DEF)
```

```python
#Creating new columns date data
dfSome2['YEAR']=dfSome2.FECHA_FALLECIMIENTO.dt.year
dfSome2['MONTH']=dfSome2.FECHA_FALLECIMIENTO.dt.month
```
